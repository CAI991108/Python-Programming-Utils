### <div class="alert alert-info">[GRADED  TASK 0] (5%)</div>
With a given integral number n, write a program to generate a dictionary that contains (**i, i!**) such that is an integral number between 1 and n (both included). and then the program should print the dictionary.

```python
import pandas as pd
from psutil import cpu_freq

def your_dictionary(n):
  """ return integral number and its factorial number between 1 and n in the dictionary.

  >>> your_dictionary(8)
  {1: 1, 2: 2, 3: 6, 4: 24, 5: 120, 6: 720, 7: 5040, 8: 40320}
  """
  # Initialize an empty dictionary to store the numbers and their factorials
  factorial_dict = {}
  
  # Iterate over the range from 1 to n (inclusive)
  for i in range(1, n+1):
      # initialize the factorial for the current number i to 1
      factorial = 1
      
      # Calculate the factorial of i by multiplying with all numbers from 1 to i
      for j in range(1, i + 1):
          factorial *= j
          
      # Add the current number and its factorial to the dictionary    
      factorial_dict[i] = factorial
  return factorial_dict
```

### <div class="alert alert-info">[GRADED  TASK 1] (5%)</div>
Implement repeated, which takes in an iterator t and an integer k greater than 1. It returns the first value in t that appears k times in a row.

Important: Call next on t only the minimum number of times required. Assume that there is an element of t repeated at least k times in a row.



```python
def repeated(t, k):
    """Return the first value in iterator t that appears k times in a row,
    calling next on t as few times as possible.

    >>> s = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> repeated(s, 2)
    9
    >>> s2 = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> repeated(s2, 3)
    8
    >>> s = iter([3, 2, 2, 2, 1, 2, 1, 4, 4, 5, 5, 5])
    >>> repeated(s, 3)
    2
    >>> repeated(s, 3)
    5
    >>> s2 = iter([4, 1, 6, 6, 7, 7, 8, 8, 2, 2, 2, 5])
    >>> repeated(s2, 3)
    2
    """
    # Initialize the previous element to None and the count to 0
    prev = None
    count = 0

    # Iterate over the iterator t
    for current in t:
        # If the current element is the same as the previous one, increment the count
        if current == prev:
            count += 1
            # If the count reaches k, return the current element as it's repeated k times
            if count == k:
                return current
        else:
            # If the current element is different from the previous one, reset the count and update prev
            prev = current
            count = 1
```

```python
s = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
s2 = iter([4, 1, 6, 6, 7, 7, 8, 8, 2, 2, 2, 5])
print(repeated(s, 2))
print(repeated(s2, 3))
```

### <div class="alert alert-info">[GRADED  TASK 2] (5%)</div>
Implement sums, which takes two positive integers n and k. It returns a list of lists containing all
the ways that a list of k positive integers can sum to n. Results can appear in any order.

```python
def sums(n, k):
  """Return the ways in which K positive integers can sum to N.
  >>> sums(2, 2)
  [[1, 1]]
  >>> sums(2, 3)
  []
  >>> sums(4, 2)
  [[3, 1], [2, 2], [1, 3]]
  >>> sums(5, 3)
  [[3, 1, 1], [2, 2, 1], [1, 3, 1], [2, 1, 2], [1, 2, 2], [1, 1, 3]]
  """
  # Base case: if k is 1, return a list with the single number n
  if k == 1:
      return [[n]] if n > 0 else []
      
  # Initialize an empty list to store the results
  result = []
    
  # Iterate from 1 to n (inclusive) to find all possible combinations
  for i in range(1, n - k + 2):
      # Recursively find the combinations for the remaining sum and (k-1) numbers
      for combination in sums(n - i, k - 1):
          # Append the current number to each combination and add to the result
          result.append([i] + combination)
  return result
```

```python
print(sums(2, 2))
print(sums(5, 3))
```

### <div class="alert alert-info">[GRADED  TASK 3] (5%)</div>
Write a recursive function that takes in a number n and determines if the digits contain two adjacent 5s. You can assume that n is at least a two-digit number.



```python
def double_fives(n):
    """ Returns whether n has two digits in row that
    are the number 5. Assume n has at least two digits in it.

    >>> double_fives(1255)
    True
    >>> double_fives(550)
    True
    >>> double_fives(535535)
    True
    >>> double_fives(254652)
    False
    >>> double_fives(553135)
    True
    >>> double_fives(78)
    False
    """
    # Convert the number to a string to easily access each digit
    str_n = str(n)
    
    # Base case: if the number has only two digits, check if they are both 5
    if len(str_n) == 2:
        return str_n[0] == '5' and str_n[1] == '5'
    
    # Recursive case: check the last two digits and the rest of the number
    last_two = str_n[-2:]  # Get the last two digits
    rest = str_n[:-2]      # Get the rest of the number
    
    # Check if the last two digits are both 5
    if last_two == '55':
        return True
    
    # If not, convert the rest of the number to an integer and recurse
    return double_fives(int(rest))
```

Implement intersection(lst_of_lsts), which takes a list of lists and returns a list of elements
that appear in all the lists in lst_of_lsts. If no number appears in all of the lists, return the empty list.
You may assume that lst_of_lsts contains at least one list.
Hint: recall that you can check if an element is in a list with the in operator:

##### >>> x = [1, 2, 3, 4]
##### >>> 4 in x
##### True
##### >>> 5 in x
##### False
##### >>>

```python
def intersection(lst_of_lsts):
  """Returns a list of elements that appear in every list in
  lst_of_lsts.
  >>> lsts1 = [[1, 2, 3], [1, 3, 5]]
  >>> intersection(lsts1)
  [1, 3]
  >>> lsts2 = [[1, 4, 2, 6], [7, 2, 4], [4, 4]]
  >>> intersection(lsts2)
  [4]
  >>> lsts3 = [[1, 2, 3], [4, 5], [7, 8, 9, 10]]
  >>> intersection(lsts3) # No number appears in all lists
  []
  """
  # Start with the first list as the initial set of common elements
  common_elements = set(lst_of_lsts[0])
    
  # Iterate over the rest of the lists
  for lst in lst_of_lsts[1:]:
      # Update the common elements to only include those that are in the current list
      common_elements &= set(lst)
    
  # Convert the set back to a list before returning
  return list(common_elements)
```

A number n contains a sandwich if a digit in n is surrounded by two identical digits. For example, the
number 242 contains a sandwich because 4 is surrounded by 2 on both sides. 1242 also contains a sandwich,
while 12532 does not contain a sandwich.
Implement the sandwich(n) function, which takes in a nonnegative integer n. It returns True if n contains
a sandwich and False otherwise. If n has fewer than three digits, it cannot contain a sandwich.

```python
def sandwich(n):
  """Returns True if n contains a sandwich and False otherwise
  >>> sandwich(416263) # 626
  True
  >>> sandwich(5050) # 505 or 050
  True
  >>> sandwich(4441) # 444
  True
  >>> sandwich(1231)
  False
  >>> sandwich(55)
  False
  >>> sandwich(4456)
  False
  """
  # Convert the integer to a string to easily access each digit
  str_n = str(n)
    
  # If the number has fewer than 3 digits, it cannot contain a sandwich
  if len(str_n) < 3:
      return False
    
  # Iterate over the string, checking each digit and its neighbors
  for i in range(1, len(str_n) - 1):
      # Check if the current digit has the same values for the one before and after it
      if str_n[i - 1] == str_n[i + 1]:
          # If so, return True, indicating a sandwich is found
          return True
    
  # If no sandwich is found after checking all digits, return False
  return False
```

```python
print(sandwich(416263))
print(sandwich(5050))
print(sandwich(4441))
print(sandwich(4456))
```

- 95% VaR (Value at Risk): 
    - With 95% confidence, the portfolio is not expected to lose more than 0.9% in a single day. 
    - In other words, there is a 5% chance that the daily loss could exceed this threshold.

- 95% CVaR (Conditional Value at Risk): 
    - CVaR, also known as Expected Shortfall (ES), is the expected loss given that a loss beyond the VaR has occurred.
    - It provides a more comprehensive view of the risk by looking at the average loss in the tail of the distribution. 
    - This value is lower than the VaR, which is expected since CVaR considers the average of the worst 5% of outcomes.

- Comparison of VaR and CVaR: 
    - The CVaR is more conservative than the VaR because the magnitude of losses exceeding the VaR threshold.
    - The CVaR is about twice the VaR (-1.6% compared to -0.9%), indicating expected significance when losses exceed the VaR.

- Risk Management Implications: 
    - While VaR gives a point estimate of potential loss, CVaR provides additional insight into the potential severity of losses. 
    - This could be particularly important for risk-averse investors or institutions that need to plan for worst-case scenarios.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
# Ignore specific warning
warnings.filterwarnings("ignore", category=UserWarning)

# Load the data
data = pd.read_csv('Score.csv')

# Select only two features for simplicity
features = ['math score', 'reading score']
X = data[features]
y = data['test preparation course']

# Convert the target variable to numerical values
y = y.map({'none': 0, 'completed': 1})

# Define the range of train sizes
train_sizes = np.linspace(0.1, 0.95, 18)
accuracies = []

# Train and evaluate for each train size
for train_size in train_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-train_size, random_state=42)
    
    # Train a logistic regression model
    model = LogisticRegression()
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    
    print(f"Train size: {round(train_size, 3)*100}%, Linear Classifier Performance Accuracy: {round(accuracy, 3)}")

# Create two figures with 3x3 subplots each
fig1, axes1 = plt.subplots(3, 3, figsize=(15, 15))
fig2, axes2 = plt.subplots(3, 3, figsize=(15, 15))

# Plot decision boundaries for each model
for i, train_size in enumerate(train_sizes):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-train_size, random_state=42)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    
    # Create grid to evaluate model
    xx = np.linspace(X[features[0]].min(), X[features[0]].max(), 30)
    yy = np.linspace(X[features[1]].min(), X[features[1]].max(), 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.predict(xy).reshape(XX.shape)
    
    # Determine which figure and axis to use
    if i < 9:
        ax = axes1[i // 3, i % 3]
    else:
        ax = axes2[(i - 9) // 3, (i - 9) % 3]
    
    # Plot decision boundary and contour
    sns.scatterplot(x=features[0], y=features[1], hue=y, data=X, palette={0: 'blue', 1: 'orange'}, legend=False, alpha=0.6, ax=ax)
    ax.contour(XX, YY, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['-'])
    ax.set_title(f'Train Size: {train_size*100:.1f}%')
    ax.set_xlabel(features[0])
    ax.set_ylabel(features[1])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data loading and filtering
data = pd.read_excel('Sample - Superstore.xls')
furniture_data = data[data['Category'] == 'Furniture']

# Group by Order Date
grouped_data = furniture_data.groupby('Order Date')['Sales'].sum().reset_index()

# Resample by month
grouped_data['Order Date'] = pd.to_datetime(grouped_data['Order Date'])
grouped_data.set_index('Order Date', inplace=True)
monthly_data = grouped_data.resample('M').sum()

# Plotting and visualization
plt.figure(figsize=(14, 7))
plt.plot(grouped_data, label='Raw Data')
plt.plot(monthly_data, label='Monthly Resampled Data', color='orange')
plt.title('Sales Trend and Seasonality')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

- Trend Analysis:
    - The monthly resampled data shows a general upward trend in sales over the observed period. 
    - There are several noticeable peaks and troughs, indicating periods of higher and lower sales activity.
- Seasonality Analysis:
    - The recurring patterns suggests the increased sales that could be associated with seasonal factors 
        - such as holidays, end-of-year spending, or other seasonal economic activities.
    - There are several distinct peaks each year, which may correspond to high-demand seasons. 
        - For instance, there's a noticeable peak towards the end of each year (Black Friday Sales and Christmas Sales)
    - The dips in sales also appear to reoccur annually, indicating periods of lower demand 
- Additional Observations:
    - The most significant increases in sales are observed in 2015 and 2017, with 2017 showing the highest peak.
    - The early months of each year tend to start with a lower sales figure, with an increase as the year progresses.
    - The raw daily sales data shows a lot of noise, and it's less useful for identifying long-term trends and seasonality.

#### NB: I'm afraid I might have some miss-understandings of the upsampling and downsampling tasks, so I tried two different approaches:

- Upsampling increases the number of data points, while downsampling reduces them.
- Upsampling uses backward filling, which means that new weekly data points are filled with the value of the previous monthly data point.
- Downsampling does not involve filling but rather selects the last data point of each month.
- The upsampled weekly data may not be representative for some weeks because they are filled values. 
- The downsampled monthly data may lose some of the variations from the original monthly data since it only retains the last data point of each month.

```python
# Upsample monthly data to weekly data
weekly_data = monthly_data.resample('W').bfill()

# Downsample weekly data to monthly data
downsampled_weekly_data = weekly_data.resample('M').last()

# Plot upsampled and downsampled data
plt.figure(figsize=(12, 6))
plt.plot(weekly_data.index, weekly_data['Sales'], 
         label='Upsampled Weekly Data', color='blue')
plt.plot(downsampled_weekly_data.index, downsampled_weekly_data['Sales'], 
         label='Downsampled Weekly Data', color='green')
plt.legend()
plt.tight_layout()
plt.show()
```

In the chart, the upsampled data shows more fluctuations within each month, while the downsampled data is smoother, reflecting only the last data point of each month.

```python
# Upsample monthly data to weekly frequency using interpolation
upsampled_data = monthly_data.interpolate()

# Downsample weekly data to monthly frequency
downsampled_data = weekly_data.resample('M').sum()

# Plot both upsampled and downsampled data
plt.figure(figsize=(10, 6))
plt.plot(upsampled_data.index, upsampled_data['Sales'], label='Upsampled Monthly to Weekly', color='blue')
plt.plot(downsampled_data.index, downsampled_data['Sales'], label='Downsampled Weekly to Monthly', color='orange')
plt.title('Sales Data - Upsampled and Downsampled')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

- The upsampled data provides a more detailed view of the sales data, showing weekly variations that are not visible in the downsampled data.
- The downsampled data provides a more aggregated view, which can be useful for identifying longer-term trends and making high-level decisions.

```python
# Shift monthly data by 1 month
shifted_data = monthly_data.shift(1)

# Plot the original and shifted data
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['Sales'], label='Original Monthly Data')
plt.plot(shifted_data.index, shifted_data['Sales'], label='Shifted Monthly Data', linestyle='--')
plt.title('Time Shifting')
plt.legend()
plt.show()
```

- The main difference is the introduction of a time lag:
- By shifting the data, each month's data is compared to the previous month's data.
- Useful for analyzing seasonal trends or cyclical patterns of the same month in different years with similar patterns.
- Visually, the shifted data lags behind the original data by one month. 
- This helps identify consistent patterns that occur a month after a particular event or season.
- Forecast future sales by comparing the current month's data with the previous month's sales (assuming similar conditions)
